```python
# %% [markdown]
# –õ–∞–±–æ—Ä–∞—Ç–æ—Ä–Ω–∞ —Ä–æ–±–æ—Ç–∞ 5: –†–µ–∞–ª—ñ–∑–∞—Ü—ñ—è –±–∞–≥–∞—Ç–æ—à–∞—Ä–æ–≤–æ–≥–æ –ø–µ—Ä—Ü–µ–ø—Ç—Ä–æ–Ω—É ¬´–∑ –Ω—É–ª—è¬ª (PyTorch –±–µ–∑ autograd)
# —ñ –µ—Ç–∞–ª–æ–Ω–Ω–∞ —Ä–µ–∞–ª—ñ–∑–∞—Ü—ñ—è –Ω–∞ nn.Module. –î–∞—Ç–∞—Å–µ—Ç Wine Quality.
# 
# üîß –ü–ª–∞–Ω:
# 1) –ó–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è —Ç–∞ –ø—Ä–µ–ø—Ä–æ—Ü–µ—Å—ñ–Ω–≥ –¥–∞–Ω–∏—Ö
# 2) –ê–∫—Ç–∏–≤–∞—Ü—ñ—ó (forward + derivative)
# 3) –§—É–Ω–∫—Ü—ñ—è –≤—Ç—Ä–∞—Ç (softmax + cross-entropy) —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–∞
# 4) –í–ª–∞—Å–Ω–∏–π MLP (—ñ–Ω—ñ—Ü—ñ–∞–ª—ñ–∑–∞—Ü—ñ—è, forward, backward, –æ–Ω–æ–≤–ª–µ–Ω–Ω—è –≤–∞–≥)
# 5) –ù–∞–≤—á–∞–Ω–Ω—è/–≤–∞–ª—ñ–¥–∞—Ü—ñ—è/—Ç–µ—Å—Ç—É–≤–∞–Ω–Ω—è –≤–ª–∞—Å–Ω–æ—ó –º–æ–¥–µ–ª—ñ
# 6) –ï—Ç–∞–ª–æ–Ω–Ω–∞ —Ä–µ–∞–ª—ñ–∑–∞—Ü—ñ—è –Ω–∞ PyTorch (nn.Module)
# 7) –ü–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ–≤ (–º–µ—Ç—Ä–∏–∫–∏, –º–∞—Ç—Ä–∏—Ü—ñ –ø–ª—É—Ç–∞–Ω–Ω—è, –≥—Ä–∞—Ñ—ñ–∫–∏)
```

```python
# %% –Ü–º–ø–æ—Ä—Ç–∏ —Ç–∞ –Ω–∞–ª–∞—à—Ç—É–≤–∞–Ω–Ω—è
import os
import math
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from typing import List, Tuple, Dict

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

import torch
import torch.nn as nn

# –í—ñ–¥—Ç–≤–æ—Ä—é–≤–∞–Ω—ñ—Å—Ç—å
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)

device = torch.device("cpu")  # –¥–ª—è —Ü—ñ—î—ó —Ä–æ–±–æ—Ç–∏ –¥–æ—Å—Ç–∞—Ç–Ω—å–æ CPU
print("Device:", device)
```

```python
# %% –ó–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è —Ç–∞ –ø—Ä–µ–ø—Ä–æ—Ü–µ—Å—ñ–Ω–≥ –¥–∞–Ω–∏—Ö
# –í–∞—à —à–ª—è—Ö –¥–æ CSV
DATA_PATH = r"D:\4K\–ú–∞—à–∏–Ω–Ω–µ –Ω–∞–≤—á–∞–Ω–Ω—è\Labs\5\WineQT.csv"

# 1) –ó–∞–≤–∞–Ω—Ç–∞–∂–∏—Ç–∏ CSV
df = pd.read_csv(DATA_PATH)

# 2) –í—ñ–¥–æ–∫—Ä–µ–º–∏—Ç–∏ –æ–∑–Ω–∞–∫–∏ X —Ç–∞ –º—ñ—Ç–∫—É y
#    –£ –¥–∞—Ç–∞—Å–µ—Ç—ñ —î –∫–æ–ª–æ–Ω–∫–∏ –æ–∑–Ω–∞–∫ —ñ —Ü—ñ–ª—å 'quality'. –¢–∞–∫–æ–∂ —î 'Id', —è–∫—É –≤–∏–¥–∞–ª—è—î–º–æ.
assert 'quality' in df.columns, "–£ CSV –ø–æ–≤–∏–Ω–Ω–∞ –±—É—Ç–∏ –∫–æ–ª–æ–Ω–∫–∞ 'quality'"
X = df.drop(columns=['quality', 'Id'], errors='ignore').copy()
y = df['quality'].copy()

# 3) –ü–µ—Ä–µ—Ç–≤–æ—Ä–∏–º–æ –∫–ª–∞—Å–∏ –¥–æ –ø–æ—Å–ª—ñ–¥–æ–≤–Ω–∏—Ö —ñ–Ω–¥–µ–∫—Å—ñ–≤ 0..K-1, –∞–ª–µ –∑–±–µ—Ä–µ–∂–µ–º–æ –º–∞–ø—ñ–Ω–≥
unique_labels = np.sort(y.unique())
label_to_index = {lab: i for i, lab in enumerate(unique_labels)}
index_to_label = {i: lab for lab, i in label_to_index.items()}
y_idx = y.map(label_to_index).values

# 4) Train/Val/Test split (70/15/15)
X_trainval, X_test, y_trainval, y_test = train_test_split(
    X.values, y_idx, test_size=0.15, random_state=SEED, stratify=y_idx
)
X_train, X_val, y_train, y_val = train_test_split(
    X_trainval, y_trainval, test_size=0.1765, random_state=SEED, stratify=y_trainval
)  # 0.1765 * 0.85 ‚âà 0.15

# 5) –ù–æ—Ä–º–∞–ª—ñ–∑–∞—Ü—ñ—è –æ–∑–Ω–∞–∫ (StandardScaler –Ω–∞ train, –ø–æ—Ç—ñ–º transform –Ω–∞ —ñ–Ω—à—ñ)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val   = scaler.transform(X_val)
X_test  = scaler.transform(X_test)

# 6) –£ —Ç–µ–Ω–∑–æ—Ä–∏ PyTorch
X_train_t = torch.tensor(X_train, dtype=torch.float32, device=device)
y_train_t = torch.tensor(y_train, dtype=torch.long, device=device)
X_val_t   = torch.tensor(X_val,   dtype=torch.float32, device=device)
y_val_t   = torch.tensor(y_val,   dtype=torch.long, device=device)
X_test_t  = torch.tensor(X_test,  dtype=torch.float32, device=device)
y_test_t  = torch.tensor(y_test,  dtype=torch.long, device=device)

input_dim  = X_train_t.shape[1]
num_classes = len(unique_labels)

input_dim, num_classes, unique_labels
```

```python
# %% –ê–∫—Ç–∏–≤–∞—Ü—ñ–π–Ω—ñ —Ñ—É–Ω–∫—Ü—ñ—ó (forward + derivative)
class ActivationFunctions:
    @staticmethod
    def relu(x: torch.Tensor, derivative: bool = False) -> torch.Tensor:
        if derivative:
            return (x > 0).to(x.dtype)
        return torch.clamp_min(x, 0.0)

    @staticmethod
    def tanh(x: torch.Tensor, derivative: bool = False) -> torch.Tensor:
        if derivative:
            t = torch.tanh(x)
            return 1.0 - t * t
        return torch.tanh(x)

    @staticmethod
    def identity(x: torch.Tensor, derivative: bool = False) -> torch.Tensor:
        if derivative:
            return torch.ones_like(x)
        return x
```

```python
# %% –í—Ç—Ä–∞—Ç–∏: softmax + cross-entropy (—Å—Ç–∞–±—ñ–ª—å–Ω–æ) —ñ —ó—ó –ø–æ—Ö—ñ–¥–Ω–∞ –∑–∞ –ª–æ–≥—ñ—Ç–∞–º–∏
class LossFunctions:
    @staticmethod
    def softmax_logits_to_probs(logits: torch.Tensor) -> torch.Tensor:
        # —Å—Ç–∞–±—ñ–ª—å–Ω–∏–π softmax (LSE)
        # logits: (B, C)
        z = logits - logits.max(dim=1, keepdim=True).values
        exp_z = torch.exp(z)
        probs = exp_z / exp_z.sum(dim=1, keepdim=True)
        return probs

    @staticmethod
    def cross_entropy_from_logits(logits: torch.Tensor, y_true: torch.Tensor) -> torch.Tensor:
        # logits: (B, C), y_true: (B,) –∫–ª–∞—Å-—ñ–Ω–¥–µ–∫—Å
        z = logits - logits.max(dim=1, keepdim=True).values
        logsumexp = torch.log(torch.exp(z).sum(dim=1))
        nll = -z[torch.arange(z.size(0)), y_true] + logsumexp
        return nll.mean()

    @staticmethod
    def grad_logits_softmax_ce(logits: torch.Tensor, y_true: torch.Tensor) -> torch.Tensor:
        # ‚àÇL/‚àÇlogits = softmax(logits) - one_hot(y)
        probs = LossFunctions.softmax_logits_to_probs(logits)
        grad = probs
        grad[torch.arange(logits.size(0)), y_true] -= 1.0
        grad = grad / logits.size(0)  # —É—Å–µ—Ä–µ–¥–Ω–∏–º–æ –ø–æ –±–∞—Ç—á—É
        return grad
```

```python
# %% –Ü–Ω—ñ—Ü—ñ–∞–ª—ñ–∑–∞—Ü—ñ—è –≤–∞–≥ (Xavier/Glorot) —Ç–∞ —É—Ç–∏–ª—ñ—Ç–∏
def xavier_init(in_dim: int, out_dim: int) -> torch.Tensor:
    # Xavier uniform
    limit = math.sqrt(6.0 / (in_dim + out_dim))
    return torch.empty(in_dim, out_dim).uniform_(-limit, limit)

def to_batches(X: torch.Tensor, y: torch.Tensor, batch_size: int, shuffle: bool = True):
    n = X.size(0)
    indices = torch.arange(n)
    if shuffle:
        indices = indices[torch.randperm(n)]
    for start in range(0, n, batch_size):
        idx = indices[start:start+batch_size]
        yield X[idx], y[idx]
```

```python
# %% –í–ª–∞—Å–Ω–∏–π MLP –±–µ–∑ autograd: –ø–∞—Ä–∞–º–µ—Ç—Ä–∏, forward/backward, –æ–Ω–æ–≤–ª–µ–Ω–Ω—è
class CustomMLP:
    def __init__(self, layer_sizes: List[int], activations: List[str], learning_rate: float = 0.01, weight_decay: float = 0.0):
        """
        layer_sizes: [in, h1, h2, ..., out]
        activations: –Ω–∞–ø—Ä–∏–∫–ª–∞–¥ ['relu','relu','identity'] (–¥–æ–≤–∂–∏–Ω–∞ = len(layer_sizes)-1)
        """
        assert len(activations) == len(layer_sizes) - 1, "–ö—ñ–ª—å–∫—ñ—Å—Ç—å –∞–∫—Ç–∏–≤–∞—Ü—ñ–π –º–∞—î –≤—ñ–¥–ø–æ–≤—ñ–¥–∞—Ç–∏ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ —à–∞—Ä—ñ–≤-–ª—ñ–Ω—ñ–π–æ–∫"

        self.layer_sizes = layer_sizes
        self.activations = activations
        self.lr = learning_rate
        self.wd = weight_decay  # L2

        # –ü–∞—Ä–∞–º–µ—Ç—Ä–∏
        self.params: Dict[str, torch.Tensor] = {}
        for i in range(1, len(layer_sizes)):
            in_d, out_d = layer_sizes[i-1], layer_sizes[i]
            W = xavier_init(in_d, out_d)
            b = torch.zeros(1, out_d)
            self.params[f"W{i}"] = W.to(torch.float32)
            self.params[f"b{i}"] = b.to(torch.float32)

        # –ì—Ä–∞–¥—ñ—î–Ω—Ç–∏ (–±—É–¥—É—Ç—å –∑–∞–ø–æ–≤–Ω—é–≤–∞—Ç–∏—Å—å —É backward)
        self.grads: Dict[str, torch.Tensor] = {}

    def _act(self, name: str, x: torch.Tensor, derivative: bool = False) -> torch.Tensor:
        if name == 'relu':
            return ActivationFunctions.relu(x, derivative)
        if name == 'tanh':
            return ActivationFunctions.tanh(x, derivative)
        if name == 'identity':
            return ActivationFunctions.identity(x, derivative)
        raise ValueError(f"–ù–µ–≤—ñ–¥–æ–º–∞ –∞–∫—Ç–∏–≤–∞—Ü—ñ—è: {name}")

    def forward(self, X: torch.Tensor) -> Tuple[torch.Tensor, Dict[str, torch.Tensor]]:
        """
        X: (B, in_dim)
        –ü–æ–≤–µ—Ä—Ç–∞—î logits —Ç–∞ cache (–¥–ª—è backprop).
        """
        cache = {}
        A = X  # (B, in)
        cache["A0"] = A

        L = len(self.layer_sizes) - 1
        for i in range(1, L + 1):
            W = self.params[f"W{i}"]
            b = self.params[f"b{i}"]
            Z = A @ W + b  # (B, out_i)
            cache[f"Z{i}"] = Z
            act_name = self.activations[i - 1]
            A = self._act(act_name, Z, derivative=False)
            cache[f"A{i}"] = A

        logits = cache[f"Z{L}"]  # –æ—Å—Ç–∞–Ω–Ω—ñ–π –ª—ñ–Ω—ñ–π–Ω–∏–π –≤–∏—Ö—ñ–¥ ‚Äî –¥–æ softmax
        return logits, cache

    def backward(self, cache: Dict[str, torch.Tensor], y_true: torch.Tensor):
        """
        –û–±—á–∏—Å–ª—é—î –≥—Ä–∞–¥—ñ—î–Ω—Ç–∏ –ø–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞—Ö —É self.grads.
        –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î –ø–æ—Ö—ñ–¥–Ω—É (softmax+CE) –∑–∞ –ª–æ–≥—ñ—Ç–∞–º–∏ —è–∫ dZ_L.
        """
        L = len(self.layer_sizes) - 1

        # dZ_L
        ZL = cache[f"Z{L}"]
        dZ = LossFunctions.grad_logits_softmax_ce(ZL, y_true)  # (B, out_L)

        self.grads = {}
        for i in range(L, 0, -1):
            A_prev = cache[f"A{i-1}"]  # (B, in_i)
            W = self.params[f"W{i}"]

            # dW_i = A_prev^T @ dZ_i
            dW = A_prev.T @ dZ  # (in_i, out_i)
            db = dZ.sum(dim=0, keepdim=True)  # (1, out_i)

            # L2 —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü—ñ—è (weight decay —É –≥—Ä–∞–¥—ñ—î–Ω—Ç—ñ)
            if self.wd > 0.0:
                dW = dW + self.wd * W

            self.grads[f"dW{i}"] = dW
            self.grads[f"db{i}"] = db

            if i > 1:
                # dA_prev = dZ_i @ W_i^T
                dA_prev = dZ @ W.T  # (B, in_i)
                Zi_1 = cache[f"Z{i-1}"]
                act_name_prev = self.activations[i - 2]
                dZ = dA_prev * self._act(act_name_prev, Zi_1, derivative=True)

    def step(self):
        """–û–Ω–æ–≤–ª–µ–Ω–Ω—è –≤–∞–≥ SGD/mini-batch."""
        L = len(self.layer_sizes) - 1
        for i in range(1, L + 1):
            self.params[f"W{i}"] = self.params[f"W{i}"] - self.lr * self.grads[f"dW{i}"]
            self.params[f"b{i}"] = self.params[f"b{i}"] - self.lr * self.grads[f"db{i}"]

    def predict(self, X: torch.Tensor) -> torch.Tensor:
        logits, _ = self.forward(X)
        return logits.argmax(dim=1)
```

```python
# %% –ù–∞–≤—á–∞–Ω–Ω—è –≤–ª–∞—Å–Ω–æ—ó –º–æ–¥–µ–ª—ñ (–±–µ–∑ autograd)
def evaluate_model(model: CustomMLP, X: torch.Tensor, y: torch.Tensor) -> Tuple[float, float]:
    logits, _ = model.forward(X)
    loss = LossFunctions.cross_entropy_from_logits(logits, y).item()
    preds = logits.argmax(dim=1)
    acc = (preds == y).float().mean().item()
    return loss, acc

def train_custom(
    model: CustomMLP,
    X_train: torch.Tensor, y_train: torch.Tensor,
    X_val: torch.Tensor,   y_val: torch.Tensor,
    epochs: int = 200,
    batch_size: int = 64,
    verbose: bool = True,
    early_stopping_patience: int = 20
):
    history = {"train_loss": [], "val_loss": [], "train_acc": [], "val_acc": []}
    best_val = float("inf")
    best_state = None
    patience = early_stopping_patience

    for epoch in range(1, epochs + 1):
        # --- train one epoch ---
        model_train_loss = 0.0
        model_train_correct = 0
        n_train = X_train.size(0)

        for xb, yb in to_batches(X_train, y_train, batch_size, shuffle=True):
            logits, cache = model.forward(xb)
            loss = LossFunctions.cross_entropy_from_logits(logits, yb)

            # backward (–∑–∞–ø–∏—à–µ –≥—Ä–∞–¥—ñ—î–Ω—Ç–∏ –≤ model.grads)
            model.backward(cache, yb)

            # step
            model.step()

            # –∑–±—ñ—Ä —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏
            with torch.no_grad():
                model_train_loss += loss.item() * xb.size(0)
                preds = logits.argmax(dim=1)
                model_train_correct += (preds == yb).sum().item()

        train_loss = model_train_loss / n_train
        train_acc  = model_train_correct / n_train

        # --- validation ---
        val_loss, val_acc = evaluate_model(model, X_val, y_val)

        history["train_loss"].append(train_loss)
        history["val_loss"].append(val_loss)
        history["train_acc"].append(train_acc)
        history["val_acc"].append(val_acc)

        if verbose and (epoch % 10 == 0 or epoch == 1):
            print(f"Epoch {epoch:4d} | train_loss={train_loss:.4f} acc={train_acc:.4f} | val_loss={val_loss:.4f} acc={val_acc:.4f}")

        # Early Stopping –∑–∞ val_loss
        if val_loss < best_val - 1e-6:
            best_val = val_loss
            best_state = {k: v.clone() for k, v in model.params.items()}
            patience = early_stopping_patience
        else:
            patience -= 1
            if patience == 0:
                if verbose:
                    print(f"‚èπÔ∏è Early stopping at epoch {epoch}. Best val_loss={best_val:.4f}")
                break

    # –í—ñ–¥–Ω–æ–≤–ª—é—î–º–æ –Ω–∞–π–∫—Ä–∞—â—ñ –≤–∞–≥–∏
    if best_state is not None:
        for k in model.params:
            model.params[k] = best_state[k].clone()

    return history
```

```python
# %% –ó–∞–ø—É—Å–∫ –Ω–∞–≤—á–∞–Ω–Ω—è –≤–ª–∞—Å–Ω–æ—ó –º–æ–¥–µ–ª—ñ
hidden = [64, 32]
layer_sizes = [input_dim] + hidden + [num_classes]
activations = ['relu', 'relu', 'identity']  # –æ—Å—Ç–∞–Ω–Ω—ñ–π –ª—ñ–Ω—ñ–π–Ω–∏–π, softmax –≤—Å–µ—Ä–µ–¥–∏–Ω—ñ –≤—Ç—Ä–∞—Ç

custom_model = CustomMLP(
    layer_sizes=layer_sizes,
    activations=activations,
    learning_rate=0.01,
    weight_decay=1e-4  # –ª–µ–≥–∫–∞ L2
)

history_custom = train_custom(
    custom_model,
    X_train_t, y_train_t,
    X_val_t, y_val_t,
    epochs=300,
    batch_size=64,
    verbose=True,
    early_stopping_patience=30
)

test_loss_custom, test_acc_custom = evaluate_model(custom_model, X_test_t, y_test_t)
print(f"[Custom] Test loss={test_loss_custom:.4f} | Test acc={test_acc_custom:.4f}")
```

```python
# %% –û—Ü—ñ–Ω–∫–∞ –≤–ª–∞—Å–Ω–æ—ó –º–æ–¥–µ–ª—ñ: –º–∞—Ç—Ä–∏—Ü—è –ø–ª—É—Ç–∞–Ω–Ω—è —Ç–∞ –∑–≤—ñ—Ç
with torch.no_grad():
    logits_test, _ = custom_model.forward(X_test_t)
    y_pred_custom = logits_test.argmax(dim=1).cpu().numpy()

acc_custom = accuracy_score(y_test, y_pred_custom)
cm_custom  = confusion_matrix(y_test, y_pred_custom)
report_custom = classification_report(
    y_test, y_pred_custom,
    target_names=[f"class_{lab}" for lab in unique_labels],
    digits=4
)

print(f"[Custom] Test Accuracy: {acc_custom:.4f}")
print("Confusion matrix (custom):\n", cm_custom)
print("\nClassification report (custom):\n", report_custom)
```

```python
# %% –ï—Ç–∞–ª–æ–Ω: PyTorch nn.Module (–¥–ª—è –ø–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è)
class PyTorchMLP(nn.Module):
    def __init__(self, input_size: int, hidden_sizes: List[int], output_size: int):
        super().__init__()
        layers: List[nn.Module] = []
        prev = input_size
        for h in hidden_sizes:
            layers.append(nn.Linear(prev, h))
            layers.append(nn.ReLU())
            prev = h
        layers.append(nn.Linear(prev, output_size))
        self.net = nn.Sequential(*layers)

    def forward(self, x):
        return self.net(x)  # logits

def train_pytorch_model(
    model: nn.Module,
    X_train: torch.Tensor, y_train: torch.Tensor,
    X_val: torch.Tensor,   y_val: torch.Tensor,
    epochs: int = 300,
    batch_size: int = 64,
    lr: float = 1e-3,
    weight_decay: float = 1e-4,
    verbose: bool = True,
    early_stopping_patience: int = 30
):
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=weight_decay)

    history = {"train_loss": [], "val_loss": [], "train_acc": [], "val_acc": []}
    best_val = float("inf")
    best_state = None
    patience = early_stopping_patience

    for epoch in range(1, epochs + 1):
        model.train()
        train_loss_sum = 0.0
        train_correct = 0
        n_train = X_train.size(0)

        # –±–∞—Ç—á—ñ
        idx = torch.randperm(n_train)
        for start in range(0, n_train, batch_size):
            sel = idx[start:start+batch_size]
            xb, yb = X_train[sel], y_train[sel]

            optimizer.zero_grad()
            logits = model(xb)
            loss = criterion(logits, yb)
            loss.backward()
            optimizer.step()

            train_loss_sum += loss.item() * xb.size(0)
            preds = logits.argmax(dim=1)
            train_correct += (preds == yb).sum().item()

        train_loss = train_loss_sum / n_train
        train_acc  = train_correct  / n_train

        # –≤–∞–ª—ñ–¥–∞—Ü—ñ—è
        model.eval()
        with torch.no_grad():
            logits = model(X_val)
            val_loss = criterion(logits, y_val).item()
            val_acc  = (logits.argmax(dim=1) == y_val).float().mean().item()

        history["train_loss"].append(train_loss)
        history["val_loss"].append(val_loss)
        history["train_acc"].append(train_acc)
        history["val_acc"].append(val_acc)

        if verbose and (epoch % 10 == 0 or epoch == 1):
            print(f"[PT] Epoch {epoch:4d} | train_loss={train_loss:.4f} acc={train_acc:.4f} | val_loss={val_loss:.4f} acc={val_acc:.4f}")

        # Early Stopping
        if val_loss < best_val - 1e-6:
            best_val = val_loss
            best_state = {k: v.detach().clone() for k, v in model.state_dict().items()}
            patience = early_stopping_patience
        else:
            patience -= 1
            if patience == 0:
                if verbose:
                    print(f"[PT] ‚èπÔ∏è Early stopping at epoch {epoch}. Best val_loss={best_val:.4f}")
                break

    if best_state is not None:
        model.load_state_dict(best_state)

    return history
```

```python
# %% –ù–∞–≤—á–∞–Ω–Ω—è PyTorch-–º–æ–¥–µ–ª—ñ
pt_model = PyTorchMLP(input_size=input_dim, hidden_sizes=hidden, output_size=num_classes).to(device)

history_pt = train_pytorch_model(
    pt_model,
    X_train_t, y_train_t,
    X_val_t, y_val_t,
    epochs=300,
    batch_size=64,
    lr=1e-3,
    weight_decay=1e-4,
    verbose=True,
    early_stopping_patience=30
)

# –û—Ü—ñ–Ω–∫–∞ –Ω–∞ —Ç–µ—Å—Ç—ñ
pt_model.eval()
with torch.no_grad():
    logits_test_pt = pt_model(X_test_t)
    y_pred_pt = logits_test_pt.argmax(dim=1).cpu().numpy()
    test_loss_pt = nn.CrossEntropyLoss()(logits_test_pt, y_test_t).item()
    test_acc_pt  = (logits_test_pt.argmax(dim=1) == y_test_t).float().mean().item()

print(f"[PyTorch] Test loss={test_loss_pt:.4f} | Test acc={test_acc_pt:.4f}")
```

```python
# %% –ü–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è —ñ—Å—Ç–æ—Ä—ñ—ó –Ω–∞–≤—á–∞–Ω–Ω—è (Custom vs PyTorch)
plt.figure(figsize=(10,4))
plt.plot(history_custom["val_loss"], label="Custom val loss")
plt.plot(history_pt["val_loss"],     label="PyTorch val loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Validation Loss: Custom vs PyTorch")
plt.legend()
plt.show()

plt.figure(figsize=(10,4))
plt.plot(history_custom["val_acc"], label="Custom val acc")
plt.plot(history_pt["val_acc"],     label="PyTorch val acc")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Validation Accuracy: Custom vs PyTorch")
plt.legend()
plt.show()
```
