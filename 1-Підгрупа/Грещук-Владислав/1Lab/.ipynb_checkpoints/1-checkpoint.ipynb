# Практична робота: Множинна лінійна регресія методом градієнтного спуску
**Мета:** реалізувати MLR з нуля (градієнтний спуск), провести EDA і перевірити 6 припущень МНК.

Інструкція: запусти клітинки по порядку.


```python
# --- 1. Імпорт бібліотек ---
!pip install numpy pandas matplotlib seaborn scikit-learn statsmodels scipy
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_squared_error, r2_score

import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

from scipy import stats

# Встановлення стилю графіків
sns.set(style="whitegrid")
%matplotlib inline
```

```python
# --- 2. Завантаження даних ---
fn = "Student_Performance.csv"  # файл має бути в тій самій папці
if not os.path.exists(fn):
    raise FileNotFoundError(f"Файл {fn} не знайдено в поточній папці: {os.getcwd()}")

df = pd.read_csv(fn)
print("Доступні стовпці:")
print(df.columns.tolist())
print("\nПерші 5 рядків:")
display(df.head())
```

```python
# Якщо назва таргету є 'Performance Index' — використовуємо її, інакше беремо останній стовпець
target_candidates = ["Performance Index", "Performance_Index", "PerformanceIndex", "performance_index", "performance index"]
target = None
for cand in target_candidates:
    if cand in df.columns:
        target = cand
        break
if target is None:
    # беремо останній стовпець як таргет (і виводимо повідомлення)
    target = df.columns[-1]
    print(f"\nУвага: 'Performance Index' не знайдено. Автоматично обрано таргет: '{target}'")

print(f"\nОбраний таргет: {target}")
```

```python
# --- 3. EDA ---
print("\nОпис датафрейму:")
display(df.info())
display(df.describe(include='all').T)

# Пропущені значення
print("\nПропущені значення по стовпцях:")
print(df.isnull().sum())

# Матриця кореляції (тільки числові)
num_df = df.select_dtypes(include=[np.number])
if num_df.shape[1] > 1:
    plt.figure(figsize=(10, 8))
    sns.heatmap(num_df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
    plt.title("Кореляційна матриця (числові змінні)")
    plt.show()
else:
    print("У наборі мало числових змінних для побудови heatmap.")
```

```python
# --- 4. Обробка категоріальних змінних ---
# Розділимо фічі на числові та категоріальні (object або категоріальні)
X = df.drop(columns=[target])
y = df[target].values

num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = X.select_dtypes(include=['object', 'category']).columns.tolist()

print("Числові стовпці:", num_cols)
print("Категоріальні стовпці:", cat_cols)

# Простий підхід: для бінарних категорій мапимо вручну, для мультикат - one-hot
# Побудуємо ColumnTransformer
transformers = []

if cat_cols:
    # OneHotEncoder для категоріальних
    transformers.append((
        'cat',
        OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'),
        cat_cols
    ))

# Для числових залишимо як є (StandardScaler застосуємо пізніше)
if num_cols:
    transformers.append(('num', 'passthrough', num_cols))

if transformers:
    preprocessor = ColumnTransformer(transformers=transformers, remainder='drop')
    X_trans = preprocessor.fit_transform(X)

    # Формуємо імена ознак
    feature_names = []
    ohe_feature_names = []
    if cat_cols:
        ohe = preprocessor.named_transformers_['cat']
        cats = ohe.categories_
        for col, cats_list in zip(cat_cols, cats):
            # drop='first' => пропускаємо перше значення
            for val in cats_list[1:]:
                ohe_feature_names.append(f"{col}_{val}")
    feature_names = ohe_feature_names + num_cols

    X_pre = pd.DataFrame(X_trans, columns=feature_names)
else:
    X_pre = X[num_cols].copy()
    feature_names = X_pre.columns.tolist()


print("\nФінальний список ознак (features):", feature_names)
display(X_pre.head())
```

```python
# --- 5. Поділ на train/test та стандартизація ---
X_train, X_test, y_train, y_test = train_test_split(X_pre.values, y, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Додаємо стовпчик одиниць для інтерсепту
X_train_final = np.c_[np.ones(X_train_scaled.shape[0]), X_train_scaled]
X_test_final = np.c_[np.ones(X_test_scaled.shape[0]), X_test_scaled]

print("Розміри X_train_final, X_test_final:", X_train_final.shape, X_test_final.shape)
```

```python
# --- 6. Реалізація градієнтного спуску з нуля ---
def compute_cost(X, y, b):
    """
    MSE (mean squared error) for parameters b.
    X: (n_samples, n_features)
    y: (n_samples,)
    b: (n_features,) or (n_features,1)
    """
    n = len(y)
    preds = X.dot(b)
    errors = preds - y
    mse = (1.0 / n) * np.sum(errors ** 2)
    return mse

def gradient_descent(X, y, b_init=None, learning_rate=0.01, epochs=1000, tol=None, verbose=False):
    n, m = X.shape
    if b_init is None:
        b = np.zeros(m)
    else:
        b = b_init.copy()
    cost_history = []
    for i in range(epochs):
        preds = X.dot(b)
        errors = preds - y
        grad = (2.0 / n) * X.T.dot(errors)   # похідна MSE
        b = b - learning_rate * grad
        cost = compute_cost(X, y, b)
        cost_history.append(cost)
        if verbose and (i % (epochs//10 + 1) == 0):
            print(f"Iter {i+1}/{epochs}, cost={cost:.6f}")
        if tol is not None and i>0 and abs(cost_history[-2]-cost_history[-1]) < tol:
            if verbose:
                print(f"Зупинка за толерантністю на ітерації {i+1}")
            break
    return b, cost_history
```

```python
# --- 7. Навчання ---
# Налаштування гіперпараметрів
learning_rate = 0.01
epochs = 5000

b_init = np.zeros(X_train_final.shape[1])
b_final, cost_history = gradient_descent(X_train_final, y_train, b_init=b_init, learning_rate=learning_rate, epochs=epochs, verbose=True)

print("\nКоефіцієнти (b_final):")
for i, coef in enumerate(b_final):
    name = "Intercept" if i==0 else feature_names[i-1]
    print(f"{name}: {coef:.6f}")
```

```python
# Графік збіжності функції втрат
plt.figure(figsize=(8,5))
plt.plot(cost_history)
plt.xlabel("Ітерація")
plt.ylabel("MSE")
plt.title("Збіжність градієнтного спуску (MSE по ітераціях)")
plt.yscale('log')  # іноді корисно бачити в логшкалі
plt.show()
```

```python
# --- 8. Прогнози та метрики ---
y_pred_test = X_test_final.dot(b_final)
mse_test = mean_squared_error(y_test, y_pred_test)
r2 = r2_score(y_test, y_pred_test)

# Бейзлайн: завжди прогноз = середнє y_train
baseline_pred = np.full_like(y_test, np.mean(y_train))
mse_baseline = mean_squared_error(y_test, baseline_pred)

print(f"Test MSE: {mse_test:.4f}")
print(f"Test R^2: {r2:.4f}")
print(f"Baseline (predict mean) MSE: {mse_baseline:.4f}")
```

```python
# --- 9. Перевірка 6 припущень МНК -- підготовка залишків ---
residuals = y_test - y_pred_test
fitted = y_pred_test
n_test = len(y_test)
```

```python
# 1) Лінійність / графік залишків vs прогнозів
plt.figure(figsize=(7,5))
plt.scatter(fitted, residuals, alpha=0.7)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Прогнозовані значення")
plt.ylabel("Залишки (y_test - y_pred)")
plt.title("Залишки vs Прогнози — перевірка лінійності / гомоскедастичності")
plt.show()

# Додатково гистограма залишків
plt.figure(figsize=(7,4))
plt.hist(residuals, bins=25)
plt.title("Гістограма залишків")
plt.show()
```

```python
# 2) Відсутність мультиколінеарності: розрахунок VIF (потрібен DataFrame ознак без інтерсепту)
# Використаємо X_train_scaled (без одиничного стовпчика)
X_vif = pd.DataFrame(X_train_scaled, columns=feature_names)
vif_data = pd.DataFrame()
vif_data['feature'] = X_vif.columns
vif_data['VIF'] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]
display(vif_data.sort_values('VIF', ascending=False))
```

```python
# 3) Нормальність залишків
# QQ-plot
plt.figure(figsize=(6,6))
stats.probplot(residuals, dist="norm", plot=plt)
plt.title("Q-Q plot залишків")
plt.show()

# Тести нормальності
shapiro_stat, shapiro_p = stats.shapiro(residuals)
jb_stat, jb_p = stats.jarque_bera(residuals)
print(f"Shapiro-Wilk: stat={shapiro_stat:.4f}, p={shapiro_p:.4f}")
print(f"Jarque-Bera: stat={jb_stat:.4f}, p={jb_p:.4f}")
```

```python
# 4) Гомоскедастичність (тест Брейша-Пагана через statsmodels)
# Для BP тесту побудуємо OLS через statsmodels (потрібно додати інтерсепт)
X_train_sm = sm.add_constant(X_train_scaled)  # для OLS у statsmodels
ols_model = sm.OLS(y_train, X_train_sm).fit()
# Використаємо residuals від OLS-моделі на train для тесту Брєша-Пагана
from statsmodels.stats.diagnostic import het_breuschpagan
bp_test = het_breuschpagan(ols_model.resid, ols_model.model.exog)
labels = ['LM stat','LM pvalue','F stat','F pvalue']
print("Breusch-Pagan test (train residuals):")
for lab, val in zip(labels, bp_test):
    print(f"{lab}: {val:.4f}")
```

```python
# 5) Відсутність автокореляції (Durbin-Watson) — для тестових залишків
from statsmodels.stats.stattools import durbin_watson
dw = durbin_watson(residuals)
print(f"Durbin-Watson (test residuals) = {dw:.4f}  (близько 2 => відсутність автокореляції)")
```

```python
# 6) Викиди та впливові спостереження (Cook's distance) — використаємо statsmodels OLS на всьому наборі для зручності
# Побудуємо OLS на всіх даних (scaled) щоб отримати influence measures
X_sm_all = sm.add_constant(np.vstack([X_train_scaled, X_test_scaled]))
y_all = np.concatenate([y_train, y_test])
ols_all = sm.OLS(y_all, X_sm_all).fit()
influence = ols_all.get_influence()
cooks_d, p_vals = influence.cooks_distance

# Порогове значення: 4/n
n_all = X_sm_all.shape[0]
threshold = 4.0 / n_all
print(f"Порог для Cook's D = 4/n = {threshold:.6f}")
# Виведемо найбільші Cook's D
cook_df = pd.DataFrame({'cooks_d': cooks_d})
cook_df_sorted = cook_df.sort_values('cooks_d', ascending=False).head(10)
display(cook_df_sorted)

# Відзначимо, скільки спостережень перевищують поріг
num_influential = (cooks_d > threshold).sum()
print(f"Кількість впливових спостережень (Cook's D > 4/n): {num_influential}")
```

```python
# --- Додаткові графіки: залишки по порядку спостережень (щоб візуально помітити автокореляцію) ---
plt.figure(figsize=(10,4))
plt.plot(residuals, marker='o', linestyle='-', alpha=0.7)
plt.axhline(0, color='red', linestyle='--')
plt.title("Залишки по індексу (test set)")
plt.xlabel("Індекс (test examples)")
plt.ylabel("Залишки")
plt.show()
```

```python
# --- 10. Висновки (автоматичний короткий звіт) ---
print("Короткий автоматичний звіт по моделі:")
print(f"- Обрана цільова змінна: {target}")
print(f"- Кількість ознак після перетворень: {len(feature_names)}")
print(f"- Test MSE: {mse_test:.4f}")
print(f"- Test R^2: {r2:.4f}")
print(f"- Baseline MSE (predict mean): {mse_baseline:.4f}")
print()
print("Перевірка припущень (коротко):")
print(f"- Нормальність (Shapiro p={shapiro_p:.3f}, JB p={jb_p:.3f}) — p > 0.05 означає, що нормальність не відкидається.")
print(f"- Гомоскедастичність (BP тест) — див. вивід; p-value маленьке => існує підозра на гетероскедастичність.")
print(f"- Автокореляція (Durbin-Watson) = {dw:.3f} (біля 2 => ок).")
print(f"- Мультиколінеарність (VIF) — див. таблицю VIF; VIF > 5 або >10 вказує на проблеми.")
print(f"- Впливові спостереження (Cook's D > 4/n): {num_influential} (див. таблицю Cook's D).")

print("\nРекомендації (якщо є проблеми):")
print("- Якщо VIF високі: розглянути видалення/агрегацію корельованих ознак або PCA.")
print("- Якщо залишки не нормальні або є гетероскедастичність: спробувати трансформації таргету (log, sqrt) або використовувати робастні методи/взвешену регресію.")
print("- Якщо є впливові спостереження: перевірити їх на предмет помилок у даних; розглянути навчання моделі без них.")
```
